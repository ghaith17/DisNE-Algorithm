```python
import time
import numpy as np
import itertools
import math
import matplotlib.pyplot as plt






def generate_tasks(num_tasks, max_cap_num, capabilities):
    return [sorted(np.random.choice(capabilities, np.random.randint(1, max_cap_num + 1), replace=False)) for _ in range(num_tasks)]



def generate_constraints(num_robots, num_tasks, power=1, min_robot_edge=2, max_task_edge=5):
    seats = max_task_edge * num_tasks
    robot_task_indices = [[] for _ in range(num_robots)]
    task_numbersers = [0 for _ in range(num_tasks)]

    robot_task_numberss = []
    for i in range(num_robots):
        max_robot_edge = min((seats - (num_robots - 1 - i) * min_robot_edge), max_task_edge)
        min_robot_edge = min(min_robot_edge, max_robot_edge)
        robot_task_numberss.append(np.random.randint(min_robot_edge, max_robot_edge + 1))
        seats -= robot_task_numberss[i]

    task_indexes = [j for j in range(num_tasks) if task_numbersers[j] < max_task_edge]

    for i in range(num_robots):
        if any([tc == 0 for tc in task_numbersers]):
            task_prob = [(math.e ** (max_task_edge - task_numbersers[j])) ** power for j in task_indexes]
            sum_prob = sum(task_prob)
            robot_task_indices[i] = list(np.random.choice(task_indexes, min(robot_task_numberss[i], len(task_indexes)), replace=False, p=[prop / sum_prob for prop in task_prob]))
        else:
            robot_task_indices[i] = list(np.random.choice(task_indexes, min(robot_task_numberss[i], len(task_indexes)), replace=False))

        for j in robot_task_indices[i]:
            task_numbersers[j] += 1
        task_indexes = [j for j in range(num_tasks) if task_numbersers[j] < max_task_edge]

    task_robots = [[i for i in range(num_robots) if j in robot_task_indices[i]] for j in range(num_tasks)]

    return robot_task_indices, task_robots


def generate_robots(constraints, tasks, max_cap_num, capabilities, max_cap_value):
    robot_task_indices = constraints[0]
    num_robots = len(robot_task_indices)
    capability_lists = []
    contribution_lists = []

    for i in range(num_robots):
        task_capabilities = [tasks[j] for j in robot_task_indices[i]]

        capabilities_union = set(itertools.chain(*task_capabilities))
        robot_cap_num = np.random.randint(min(3, max_cap_num, len(capabilities_union)), min(len(capabilities_union), max_cap_num) + 1)

        robot_caps = set([np.random.choice(t_c) for t_c in task_capabilities])

        remaining_choices = list(capabilities_union.difference(robot_caps))
        if remaining_choices:
            update_len = max(0, robot_cap_num - len(robot_task_indices[i]))
            robot_caps.update(np.random.choice(remaining_choices, min(len(remaining_choices), update_len), replace=False))

        capability_lists.append(sorted(list(robot_caps)))

        contribution_lists.append([(np.random.randint(1, max_cap_value + 1) if c in capability_lists[i] else 0) for c in range(len(capabilities))])

    return capability_lists, contribution_lists


def random(robots, tasks, constraints, gamma=1):
    task_num = len(tasks)
    robot_num = len(robots)
    a_taskInds = constraints[0]
    
    # Randomly assign tasks to robots
    alloc = [np.random.choice(a_taskInds[i] + [task_num]) for i in range(robot_num)]
    
    # Calculate the system reward
    system_reward = get_system_reward(robots, tasks, alloc, gamma)
    
    return alloc, system_reward

def task_reward(task_requirements, robot_capabilities, gamma=1):
    if not robot_capabilities:
        return 0
    else:
        # Calculate the sum of the maximum capabilities of all robots for each task requirement
        task_reward_value = sum(max(robot[capability] for robot in robot_capabilities) for capability in task_requirements)
        
        # Multiply the sum by the gamma factor raised to the power of the number of robot capabilities
        return task_reward_value * (gamma ** len(robot_capabilities))

def get_system_reward(robots, tasks, allocations, gamma=1):
    num_tasks = len(tasks)
    
    # Calculate the sum of task rewards for all tasks
    system_reward = sum(
        task_reward(tasks[j], [robots[i] for i in range(len(robots)) if allocations[i] == j], gamma) 
        for j in range(num_tasks)
    )
    
    return system_reward


def calculate_results(robots, tasks, constraints, received_messages, round, round_over, convergence, gamma=1):
    robot_task_indices = constraints[0]
    num_robots = len(robots)
    num_tasks = len(tasks)
    
    # Calculate robot_message_sums for each robot
    robot_message_sums = [
        {
            task_key: sum(
                received_messages[j][i][0] for j in robot_task_indices[i] if j != task_key
            ) + received_messages[task_key][i][1]
            for task_key in robot_task_indices[i]
        }
        for i in range(num_robots)
    ]
    
    # Determine the allocations for each robot
    allocations = [
        max(message_sum, key=message_sum.get) if message_sum else num_tasks
        for message_sum in robot_message_sums
    ]
    
    # Calculate the system reward
    system_reward = get_system_reward(robots, tasks, allocations, gamma)
    
    return allocations, system_reward, round, round_over, convergence


def robot_contribution(competency_lists, task_requirements_lists, query_robot_index, query_task_index, members, robot_task_indices):
    if query_task_index == len(task_requirements_lists):
        return 0
    if query_task_index not in robot_task_indices[query_robot_index]:
        return 0
    current_profiles = [competency_lists[member] for member in members]
    current_reward = task_reward(task_requirements_lists[query_task_index], current_profiles, gamma=1)
    if query_robot_index in members:
        remaining_members = members[:]
        remaining_members.remove(query_robot_index)
        coalition_profiles = [competency_lists[member] for member in remaining_members]
        new_reward = task_reward(task_requirements_lists[query_task_index], coalition_profiles, gamma=1)
        return current_reward - new_reward
    else:
        additional_members = members[:]
        additional_members.append(query_robot_index)
        coalition_profiles = [competency_lists[member] for member in additional_members]
        new_reward = task_reward(task_requirements_lists[query_task_index], coalition_profiles, gamma=1)
        return new_reward - current_reward

def get_target_tasks(movement_values, robot_tasks, robot_index):
    max_movement_value = max(movement_values['robot {}'.format(robot_index)])
    target_task_list = []
    max_value_count = movement_values['robot {}'.format(robot_index)].count(max_movement_value)
    start_position = 0
    for _ in range(max_value_count):
        sublist = movement_values['robot {}'.format(robot_index)][start_position:]
        task_index = start_position + sublist.index(max_movement_value)
        actual_task = robot_tasks[robot_index][task_index]
        target_task_list.append(actual_task)
        start_position += sublist.index(max_movement_value) + 1
    return target_task_list


def accepted_tasks(instruction_received, proposal_states, robot_index):
    accepted_task_list = []
    for i, instruction in enumerate(instruction_received[f'robot {robot_index}']):
        if instruction == 1:
            actual_task = proposal_states[f'robot {robot_index}'][i]
            accepted_task_list.append(actual_task)
    return accepted_task_list

def disNE(robot_tasks, task_robots, tasks, robots, time_bound, coalition_structure=[]):
    convergence = False
    num_robots = len(robots)
    num_tasks = len(tasks)

    if not coalition_structure:
        coalition_structure = [[] for _ in range(num_tasks + 1)]
        coalition_structure[num_tasks] = list(range(num_robots))

    robot_states = [num_tasks for _ in range(num_robots)]

    received_contributions = {f'robot {i}': [0] * len(robot_tasks[i]) for i in range(num_robots)}

    round = 0
    update_tasks = range(num_tasks)
    message_passed = 0
    continue_condition = True
    start_time = time.time()

    while continue_condition:
        if time.time() - start_time >= time_bound:
            break

        round += 1

        for j in update_tasks:
            for i in coalition_structure[j]:
                robot_states[i] = j

        for j in range(num_tasks):
            if j in update_tasks:
                for i in task_robots[j]:
                    message_passed += 1
                    received_contributions[f'robot {i}'][robot_tasks[i].index(j)] = robot_contribution(robots, tasks, i, j, coalition_structure[j], robot_tasks)

        movement_values = {f'robot {i}': [received_contributions[f'robot {i}'][j] - received_contributions[f'robot {i}'][robot_tasks[i].index(robot_states[i])] for j in range(len(robot_tasks[i]))] for i in range(num_robots)}

        received_proposals = [{'from robots': [], 'proposal values': []} for _ in range(num_tasks)]
        proposal_states = {}
        proposing_robots = []
        receiving_tasks = []

        for i in range(num_robots):
            max_movement_value = max(movement_values[f'robot {i}'])
            if max_movement_value > 0:
                proposing_robots.append(i)
                target_tasks = get_target_tasks(movement_values, robot_tasks, i)
                if robot_states[i] != num_tasks:
                    target_tasks.append(robot_states[i])
                message_passed += len(target_tasks)
                proposal_states[f'robot {i}'] = target_tasks
                receiving_tasks.append(target_tasks)

                for j in target_tasks:
                    received_proposals[j]['from robots'].append(i)
                    received_proposals[j]['proposal values'].append(max_movement_value)

        if max([max(mv) for mv in movement_values.values()]) > 0:
            continue_condition = True
        else:
            continue_condition = False
            convergence = True

        if continue_condition:
            active_tasks = list(set(itertools.chain(*receiving_tasks)))

            instruction_received = {}
            for i in proposing_robots:
                instruction_received[f'robot {i}'] = [0] * len(proposal_states[f'robot {i}'])
                message_passed += len(proposal_states[f'robot {i}'])

            for j in active_tasks:
                if j == num_tasks:
                    for robot_index in received_proposals[j]['from robots']:
                        robot_index = int(robot_index)
                        task_index = proposal_states[f'robot {robot_index}'].index(j)
                        instruction_received[f'robot {robot_index}'][task_index] = 1
                else:
                    max_proposal = max(received_proposals[j]['proposal values'])
                    accepted_robot_index = received_proposals[j]['proposal values'].index(max_proposal)
                    chosen_robot = received_proposals[j]['from robots'][accepted_robot_index]
                    task_index = proposal_states[f'robot {chosen_robot}'].index(j)
                    instruction_received[f'robot {chosen_robot}'][task_index] = 1
                    update_tasks = []
        for i in proposing_robots:
            if instruction_received[f'robot {i}'].count(1) == 0:
                continue
            else:
                accepted_task_list = accepted_tasks(instruction_received, proposal_states, i)
                previous_coalition = robot_states[i]
                if previous_coalition != num_tasks:
                    previous_coalition_index = proposal_states[f'robot {i}'].index(previous_coalition)
                    if instruction_received[f'robot {i}'][previous_coalition_index] == 0:
                        continue
                    else:
                        message_passed += 1
                        accepted_task_list.remove(previous_coalition)
                if not accepted_task_list:
                    continue
                else:
                    message_passed += 1
                    chosen_task = np.random.choice(accepted_task_list)
                coalition_structure[chosen_task].append(i)
                update_tasks.append(chosen_task)
                robot_index = coalition_structure[previous_coalition].index(i)
                del coalition_structure[previous_coalition][robot_index]
                update_tasks.append(previous_coalition)

    system_reward = get_system_reward(robots, tasks, robot_states, gamma=1)
    return system_reward, round, message_passed, convergence
def compare_algorithms(num_experiments, probability, time_bound, max_task_numbers, capability_count, max_capability_task, max_capability_robot, max_capability_value, capabilities, min_robot_edge, min_task_numbers, experiment_identifier):


   # TODO: Add code to run other algorithms (DSA and BnB-FMS)
    dsa_results = [
              {'utility': 2298.85, 'message_passing': 386017.9, 'running_time': 0.037302, 'rounds': 15.0848},
              {'utility': 6896.55, 'message_passing': 2710456, 'running_time': 0.946652, 'rounds':43.3574},
              {'utility': 10919.5, 'message_passing': 5159093, 'running_time': 6.1054, 'rounds': 157.31},
              {'utility': 14367.8, 'message_passing': 10401260, 'running_time': 352.874, 'rounds': 6166.14},
              {'utility': 17816.1, 'message_passing': 10364000, 'running_time': 352.874, 'rounds': 3656.03},
              {'utility': 21264.4, 'message_passing': 10391710, 'running_time': 352.874, 'rounds': 2532.24},
              {'utility': 24712.6, 'message_passing': 10355400, 'running_time': 372.759, 'rounds': 1753.88},
              {'utility': 28735.6, 'message_passing': 10382150, 'running_time': 372.759, 'rounds': 1418.51},
              {'utility': 31609.2, 'message_passing': 10407950, 'running_time': 352.874, 'rounds': 1089.4},
              {'utility': 35632.2, 'message_passing': 10371640, 'running_time': 352.874, 'rounds': 881.255},
            
            ]

          
    bnb_fms_results = [
            {'utility': 2298.85, 'message_passing': 0, 'running_time': 10, 'rounds':471.32},
            {'utility': 6896.55, 'message_passing': 0, 'running_time': 352.874, 'rounds': 468.677},
            {'utility': 10919.5, 'message_passing': 0, 'running_time': 372.759, 'rounds': 10.6687},
            {'utility': 12069, 'message_passing': 0, 'running_time': 760.243, 'rounds': 1.73565},


          ]
      
    dsa_utilities = [r['utility'] for r in dsa_results]
    bnb_fms_utilities = [r['utility'] for r in bnb_fms_results]


    dsa_running_time= [r['running_time'] for r in dsa_results]
    bnb_fms_running_time = [r['running_time'] for r in bnb_fms_results]

    dsa_rounds= [r['rounds'] for r in dsa_results]
    bnb_fms_rounds = [r['rounds'] for r in bnb_fms_results]

    dsa_message_passing= [r['message_passing'] for r in dsa_results]
    bnb_fms_message_passing = [r['message_passing'] for r in bnb_fms_results]

    
          


    # Compare algorithms
    for num_experiment in range(0, num_experiments):
        print('Experiment:', num_experiment)
        task_numbers = min_task_numbers
        task_numberss = []
        disNE_utilities = []
        disNE_message_passing= []
        disNE_running_time= []
        disNE_round= []
        while task_numbers <= max_task_numbers:
            max_task_edge = 0.04 * task_numbers
            experiment_identifier += 1
            robot_numbers = 2 * task_numbers
            tasks = generate_tasks(task_numbers, max_capability_task, capabilities)

            constraints = generate_constraints(robot_numbers, task_numbers, 1, min_robot_edge, max_task_edge)
            # caps_lists: capabilities the robot can provide; robots: the value of the robot provide the capabilities
            caps_lists, robots = generate_robots(constraints, tasks, max_capability_robot, capabilities, max_capability_value)

            robot_tasks = constraints[0]
            task_robots = constraints[1]
            for i in range(0, robot_numbers):
                robot_tasks[i].append(task_numbers)
            task_robots.append(list(range(0, robot_numbers)))

            result = {}

            start_time = time.time()
            disNE_utility, disNE_rounds, disNE_message_passed, is_converged = disNE(robot_tasks, task_robots, tasks, robots, time_bound, [])
            end_time = time.time()
            result['disNE_utility'] = disNE_utility
            result['disNE_time'] = end_time - start_time
            result['disNE_rounds'] = disNE_rounds
            result['disNE_message_passed'] = disNE_message_passed
            result['disNE_converged'] = is_converged
            
            # Append the task_numbers and disNE_utility values to the respective lists
            task_numberss.append(task_numbers)
            disNE_utilities.append(disNE_utility)
            disNE_message_passing.append(disNE_message_passed)
            disNE_running_time.append(end_time - start_time)
            disNE_round.append(disNE_rounds)

            print( "Task numbers:",task_numbers,"\t","running time:", result['disNE_time'],"\t", 'utility:', result['disNE_utility'],"\t",
                  "rounds:", result['disNE_rounds'], "\t","message passing", result['disNE_message_passed'],"\t", "Converged?",result['disNE_converged'])
            print()

            # Increase the task_numbers
            task_numbers += 100
        num_experiment += 1
        
        #Get the task numbers for DSA and BnB-FMS results
        dsa_tasks = task_numberss[:len(dsa_utilities)]
        bnb_fms_tasks = task_numberss[:len(bnb_fms_utilities)]

        # Create a scatter plot of task_numberss vs disNE_utilities
        fig = plt.figure(figsize=(4, 4), dpi=50)
        plt.title("(a) Achieved system utilities.")
        plt.plot(task_numberss, disNE_utilities, label="DisNE",marker="o")
        plt.plot(dsa_tasks, dsa_utilities, label="DSA",marker="x")
        plt.plot(bnb_fms_tasks, bnb_fms_utilities, label="BnB-FMS",marker="s")
        plt.xlabel('Task Number')
        plt.ylabel('Utility')
        plt.legend()
        plt.show()


        # Plot Running time comparison
        fig = plt.figure(figsize=(4, 4), dpi=50)
        plt.title("(b) Running time comparison")
        plt.plot(task_numberss, disNE_running_time, label="DisNE",marker="o")
        plt.plot(dsa_tasks, dsa_running_time, label="DSA",marker="x")
        plt.plot(bnb_fms_tasks, bnb_fms_running_time, label="BnB-FMS",marker="s")
        plt.xlabel("Task Number")
        plt.ylabel("Time")
        plt.xticks([200, 400, 600, 800, 1000])
        plt.yscale('log')
        plt.yticks([0.1, 1, 10, 100, 1000], [r'$10^{-1}$', r'$10^{0}$', r'$10^{1}$', r'$10^{2}$', r'$10^{3}$'])
        plt.legend()
        plt.show()

        # Plot Execution round comparison
        fig = plt.figure(figsize=(4, 4), dpi=50)
        plt.title("(c) Execution round comparison")
        plt.plot(task_numberss, disNE_round, label="DisNE",marker="o")
        plt.plot(dsa_tasks, dsa_rounds, label="DSA",marker="x")
        plt.plot(bnb_fms_tasks, bnb_fms_rounds, label="BnB-FMS",marker="s")
        plt.xlabel("Task Number")
        plt.ylabel("rounds")
        plt.xticks([200, 400, 600, 800, 1000])
        plt.yscale('log')
        plt.yticks([ 1, 10, 100, 1000], [ r'$10^{0}$', r'$10^{1}$', r'$10^{2}$', r'$10^{3}$'])
        plt.legend()

        plt.show()



        # Plot Number of messages passed.
        fig = plt.figure(figsize=(4, 4), dpi=50)
        plt.title("(d) Number of messages passed.")
        plt.plot(task_numberss, disNE_message_passing, label="DisNE",marker="o")
        plt.plot(dsa_tasks, dsa_message_passing, label="DSA",marker="x")
        plt.xlabel("Task Number")
        plt.ylabel("message passing")
        plt.yscale('log')
        plt.yticks([1000,10000,100000,1000000,10000000], [ r'$10^{3}$',r'$10^{4}$',r'$10^{5}$',r'$10^{6}$',r'$10^{7}$'])
        plt.legend()
        plt.show()



def main():
  
  compare_algorithms(1000, 0.7, 300, 1000, 10, 10, 10, 10, list(range(0, 10)), 1, 100, 120)



if __name__ == "__main__":
    main()
```
